```python
import numpy as np

import statsmodels.api as sm

import statsmodels.formula.api as smf

import pandas as pd
import plotnine as p9
```

```python
df_40=pd.read_csv('C:/Bazy danych/Analiza i Bazy Danych/exercise.csv')
```

```python
df = df_40.head(40)
```

```python
df
```

```python

results = smf.ols('y ~ x1', data=df).fit()
wyn=results.params

```

```python
wyn
```

```python
results.summary()
```

```python
import matplotlib.pyplot as plt 
```

```python
fig1=(p9.ggplot(p9.aes(x='x1',y='y'),data=df)
      +p9.geom_jitter(width=0.1)
      +p9.geom_abline(p9.aes(intercept=wyn['Intercept'],slope=wyn['x1'])))
```

```python
fig1
```

```python

results2 = smf.ols('y ~ x2', data=df).fit()
wyn2=results2.params
```

```python
results2.summary()
```

```python
fig2=(p9.ggplot(p9.aes(x='x2',y='y'),data=df)
      +p9.geom_jitter(width=0.1)
      +p9.geom_abline(p9.aes(intercept=wyn2['Intercept'],slope=wyn2['x2'])))
```

```python
fig2
```

```python
results3 = smf.ols('y ~ x1+x2', data=df).fit()
wyn3=results3.params
df['y']=results3.predict()
df['residuals']=df['x1']-df['x2']
```

```python


fig3_res=(p9.ggplot(p9.aes(x='x1',y='residuals',color='y'),data=df)
      +p9.geom_point())
fig3_res

```

```python
fig4_res=(p9.ggplot(p9.aes(x='x2',y='residuals',color='y'),data=df)
      +p9.geom_point())
fig4_res
```

```python
beauty=pd.read_csv('C:/Bazy danych/Analiza i Bazy Danych/beauty.csv')
```

```python
beauty.head()
```

```python
beauty100=beauty.head(100)
```

```python
results8 = smf.ols('courseevaluation ~ btystdave', data=beauty100).fit()
wyn8 = results8.params
results8.summary()
```

```python
fig8=(p9.ggplot(p9.aes(x='btystdave',y='courseevaluation'),data=beauty100)
      +p9.geom_point()
      +p9.geom_abline(p9.aes(intercept=wyn8['Intercept'],slope=wyn8['btystdave'])))
fig8
```

```python

beauty_pr = beauty100.copy()
beauty_pr['predict']=results8.predict()
beauty_pr['residuals']=abs(beauty_pr['courseevaluation']-beauty_pr['predict'])
```

```python
fig9=(p9.ggplot(p9.aes(x='btystdave',y='residuals'),data=beauty_pr)
      +p9.geom_point())
fig9
```

```python
results9 = smf.ols('courseevaluation ~ btystdave + female + btystdave*female', data=beauty100).fit()
wyn9 = results9.params
results9.summary()
```

```python
fig10=(p9.ggplot(p9.aes(x='female',y='courseevaluation'),data=beauty100)
      #+p9.geom_point()
             +p9.geom_jitter(width=0.1)
      +p9.geom_abline(p9.aes(intercept=wyn9['Intercept'],slope=wyn9['female'])))
fig10
```

```python
beauty_pr2 = beauty100.copy()
beauty_pr2['predict2']=results9.predict()
beauty_pr2['residuals2']=abs(beauty_pr2['courseevaluation']-beauty_pr2['predict2'])
```

```python
fig11=(p9.ggplot(p9.aes(x='btystdave',y='residuals2'),data=beauty_pr2)
      +p9.geom_point())
fig11
```

```python
results12 = smf.ols('courseevaluation ~ btystdaveneg + profevaluation', data=beauty100).fit()
wyn12 = results12.params
results12.summary()
```

```python
fig12=(p9.ggplot(p9.aes(x='profevaluation',y='courseevaluation'),data=beauty100)
      #+p9.geom_point()
             +p9.geom_jitter(width=0.1)
      +p9.geom_abline(p9.aes(intercept=wyn12['Intercept'],slope=wyn12['profevaluation'])))
fig12
```

```python
beauty_pr3 = beauty100.copy()
beauty_pr3['predict3']=results12.predict()
beauty_pr3['residuals3']=abs(beauty_pr3['courseevaluation']-beauty_pr3['predict3'])
```

```python
fig13=(p9.ggplot(p9.aes(x='courseevaluation',y='residuals3'),data=beauty_pr3)
      +p9.geom_point())
fig13
```
