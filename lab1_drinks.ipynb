```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
df=pd.read_csv('C:/Bazy danych/Analiza i Bazy Danych/drinks.csv')
```

```python
df
```

```python
df1=df.rename(columns={"beer_servings":"beer","spirit_servings":"spirit","wine_servings":"wine","total_litres_of_pure_alcohol":"litres of alcohol"})

data_melt = pd.melt(df1,id_vars=["country" , "litres of alcohol"])

```

```python
data_melt.head()
```

```python
data=data_melt.rename(columns={"variable":"kind of alcohol","value":"quantity of alcohol"})
```

```python
data
```

```python
data_tidy = data.pivot_table(
    index=['country','litres of alcohol','kind of alcohol'], 
    values='quantity of alcohol')
```
