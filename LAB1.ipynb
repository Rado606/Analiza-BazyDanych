```python
d3=df.rename(columns={"new_sp":"un_un","new_sp_m_0-4":"m_0-4","new_sp_m_5-14":"m_5-14"})
d4=d3.rename(columns={"new_sp_m_0-14":"m_0-14","new_sp_m_15-24":"m_15-24","new_sp_m_25-34":"m_25-34"})
d5=d4.rename(columns={"new_sp_m_35-44":"m_35-44","new_sp_m_45-54":"m_45-54","new_sp_m_55-64":"m_55-64","new_sp_m_65+":"m_65+","new_sp_mu":"m_un"})                  
```

```python
d6=d5.rename(columns={"new_sp_f_0-4":"f_0-4","new_sp_f_5-14":"f_5-14"})
d7=d6.rename(columns={"new_sp_f_0-14":"f_0-14","new_sp_f_15-24":"f_15-24","new_sp_f_25-34":"f_25-34"})
data=d7.rename(columns={"new_sp_f_35-44":"f_35-44","new_sp_f_45-54":"f_45-54","new_sp_f_55-64":"f_55-64","new_sp_f_65+":"f_65+","new_sp_fu":"f_un"})      
```

```python
data.head()
```

```python
data_melt = pd.melt(data,id_vars=["iso2","year"])
```

```python
data_melt.head()
```

```python
variable_split = data_melt['variable'].str.split('_')
```

```python
variable_split.head()
```

```python
data_melt["sex"]=variable_split.str.get(0)
```

```python
data_melt["age"]=variable_split.str.get(1)
```

```python
data_melt.head()
```

```python
data_melt.drop('variable',axis=1,inplace=True)
```

```python
data_melt.head()
```

```python
data_frame = data_melt[data_melt['value']>=0]
```

```python
data_frame.head()
```

```python
data_tidy = data_frame.pivot_table(
    index=['iso2','year','sex','age'], 
    values='value')
```
